```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
dataframe = pd.read_csv("dataset_with_missing_values_duplicates_removed.csv")
dataframe.head()
```

```python
X = dataframe["text"]
y = dataframe["label"]
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
print("First 50 encoded values:", y_encoded[:50])
```

```python
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_classifier.fit(X_tfidf, y_encoded)
```

```python
def predict_if_a_news_is_fake(text):
    text_tfidf = tfidf_vectorizer.transform([text])
    prediction_result = random_forest_classifier.predict(text_tfidf)
    predicted_label = label_encoder.inverse_transform(prediction_result)
    return predicted_label[0]
```

```python
text_input = "Consider NFTs, or non-fungible tokens. These are collectibles—some might call them art—created by the same block-chain technology as cryptocurrencies like bitcoin. Jack Dorsey, the founder of Twitter, turned an image of his first tweet into an NFT in December 2020 and auctioned it off for $2.9 million a few months later. Dorsey donated the proceeds to charity, but most others have taken the money and run. When the purchaser of his NFT tried to resell it in April 2022 for $50 million, the highest bid was a paltry $280."
prediction_result = predict_if_a_news_is_fake(text_input)
print(f"Prediction result: {prediction_result}")
```
